```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

stock = 'TCS.NS'
data = yf.download(stock, start='2022-01-01', end='2024-01-01', auto_adjust=False)
print("First 5 rows of stock data:")
display(data.head())

data['Daily Return'] = data['Adj Close'].pct_change()
display(data[['Adj Close', 'Daily Return']].head())


data['SMA20'] = data['Adj Close'].rolling(window=20).mean()
data['SMA50'] = data['Adj Close'].rolling(window=50).mean()


plt.figure(figsize=(14,6))
plt.plot(data['Adj Close'], label='Adjusted Close', color='blue')
plt.plot(data['SMA20'], label='20-Day SMA', color='green')
plt.plot(data['SMA50'], label='50-Day SMA', color='red')
plt.title(f"{stock} Stock Price with 20 & 50 Day Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price (INR)")
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(12,5))
data['Daily Return'].plot(kind='hist', bins=100, color='orange')
plt.title(f"{stock} Daily Return Distribution")
plt.xlabel("Daily Return")
plt.grid(True)
plt.show()
```
